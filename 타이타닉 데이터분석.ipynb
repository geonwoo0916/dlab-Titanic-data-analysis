데이터 전처리

```python
# 켤측치 확인 True, False 확인
df_t.isna()
```

```python
# 결측시 개수를 알아봅시다
df_t.isna().sum()
```

```python
df_t['Age'] = df_t['Age'].fillna(df_t['Age'].mean())
df_t.isna().sum()
```

```python
# Cabin 컬럼 삭제하기
df_t.drop('Cabin',axis=1,inplace=True)
```

```python
df_t['Embarked'].unique()
```

```python
df_t.groupby('Embarked').count()
```

```python
df_t['Embarked'] = df_t['Embarked'].fillna('S')
df_t.isna().sum()
```

```python
# 탑승항구와 티켓요금의 관계
sns.boxplot(data=df_t, x='Embarked', y='Fare')
plt.show()
```

```python
# apply 함수를 사용해 노인 성인 어린이 추가

def AgeDiff(Age):
  if Age < 16:
    return 'child'
  elif Age < 50:
    return 'adult'
  else:
    return 'old man'

df_t['AgeDiff'] = df_t['Age'].apply(AgeDiff)
```
